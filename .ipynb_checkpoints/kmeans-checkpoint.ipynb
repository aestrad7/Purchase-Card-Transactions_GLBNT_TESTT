```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import sklearn.metrics as metrics
import sklearn.cluster as cluster

from Funciones import Explorer
```

```python
DATA_DIR="/home/grupodot/Documentos/Examen_globant/Purchase-Card-Transactions_GLBNT_TESTT/Data/"
```

```python
expl = Explorer()
df_f = expl.clean_import(DATA_DIR)
print("La dimension de la base es: ", df_f.shape)
df_f.describe()
```

```python
scaler=StandardScaler()
scaler.fit(df_f)
scaled_data=scaler.transform(df_f)
df_f2 = pd.DataFrame(data=scaled_data, columns=df_f.columns)
df_f2.describe()
```

```python
wcss = []

for i in range (1,11):
    kmeans = KMeans(n_clusters= i, max_iter= 500)
    kmeans.fit(df_f2)
    wcss.append(kmeans.inertia_)
```

```python
expl.graficador(wcss, tittle = "WCSS KMeans Points")
```

```python
# Calculamos el silhouette de cada K (se eligen 15.000 datos de muestra para representar un 30% del total de datos)

for i in range(2,8):
    labels = cluster.KMeans(n_clusters=i, init ="k-means++").fit(df_f2).labels_
    print( "Valor de Silhouette = " + str(i) + " " + 
          str(metrics.silhouette_score(df_f2, labels, metric ="euclidean", sample_size = 15000,
                random_state = 42)))
```

```python
# Generamos un modelo con 3 salidas posibles

df_kmeans3 = df_f2.copy()
clustering3 = KMeans(n_clusters = 3, max_iter = 100000)
clustering3.fit(df_f2)
df_kmeans3["KMeans_cluster"] = clustering3.labels_
```

```python
pca3=PCA(n_components=2)
pca3.fit(df_f2)
x3_pca=pca3.transform(df_f2)
```

```python
plt.figure(figsize=(8,6))
plt.scatter(x3_pca[:,0],x3_pca[:,1], c = df_kmeans3["KMeans_cluster"] )
plt.xlabel('First principle component')
plt.ylabel('Second principle component')
plt.title("Tipos de transaccion agrupadas en 3 grupos")
```

```python
df_kmeans5 = df_f2.copy()

clustering5 = KMeans(n_clusters = 5, max_iter = 100000)
clustering5.fit(df_f2)

df_kmeans5["KMeans_cluster"] = clustering5.labels_
```

```python
pca5=PCA(n_components=2)
pca5.fit(df_f2)
x5_pca=pca5.transform(df_f2)
```

```python
plt.figure(figsize=(8,6))
plt.scatter(x5_pca[:,0],x5_pca[:,1],c=df_kmeans5["KMeans_cluster"])
plt.xlabel('First principle component')
plt.ylabel('Second principle component')
plt.title("Tipos de transaccion agrupadas en 5 grupos")
```

```python
g1 = df_kmeans3[df_kmeans3["KMeans_cluster"]==0]
g2 = df_kmeans3[df_kmeans3["KMeans_cluster"]==1]
g3 = df_kmeans3[df_kmeans3["KMeans_cluster"]==2]

columnsg = df_kmeans3.columns
```

```python
plt.subplot()
for i in range(len(columnsg)-1):
    expl.comparador(columnsg[i], g1, g2, g3)
```
